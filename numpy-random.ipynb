### numpy.random.random() ---- Generate random floating point numbers
The default is to generate a random floating point number, the range is between 0.0~1.0, you can also set the size of the returned data by the 
parameter size setting;

Generate a random floating point number:

```python
import numpy
n = numpy.random.random()
print (n)
```

Set the size of the returned data by the parameter size setting:
numpy.random.random(size=(3, 2))  -- means the resule is 3 rows and 2 columns. 

```python
import numpy
n = numpy.random.random(size=(3, 2))
print (n)
```

#### numpy.random.randint()　Generate random integers

API: randint(low, high=None, size=None, dtype='l')                               
Numpy.random.randint() randomly generates an integer number, which can specify the range of this integer

```python
import numpy as np
print (np.random.randint(8))
print (np.random.randint(5, size=3))
print (np.random.randint(6, size=(3,2)))
```

### numpy.random.normal()  Gaussian distribution random number

API: normal(loc=0.0, scale=1.0, size=None)

The probability density for the Gaussian distribution is

$$p(x) = \frac{1}{\sqrt{ 2 \pi \sigma^2 }}
e^{ - \frac{ (x - \mu)^2 } {2 \sigma^2} }$$,

where $\mu$ is the mean and $\sigma$ the standard deviation. The square of the standard deviation, $\sigma^2$, is called the variance.
The function has its peak at the mean, and its “spread” increases with the standard deviation (the function reaches 0.607 times its maximum at $x + \sigma$ and $x - \sigma$). This implies that [numpy.random.normal][1] is more likely to return samples lying close to the mean, rather than those far away.

[1]: https://docs.scipy.org/doc/numpy/reference/generated/numpy.random.normal.html#numpy.random.normal

```python
import numpy
n = numpy.random.normal(loc=1, scale=9, size=1)
print (n)
```

Draw sample about distribution:

```python
# 若随机变量X服从一个数学期望为μ、标准方差为σ2的高斯分布，记为：X∼N(μ,σ2), 正态分布的期望值μ决定了其位置，其标准差σ决定了分布的幅度。
# 我们通常所说的标准正态分布是μ = 0,σ = 1的正态分布（见右图中蓝色曲线）。
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0) #Make random data predictable

mu = 100  # mean of distribution
sigma = 15  # standard deviation of distribution
x = mu + sigma * np.random.randn(437)

num_bins = 30

# the histogram of the data
n, bins, patches = plt.hist(x, num_bins, density=True)

plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) *
               np.exp( - (bins - mu)**2 / (2 * sigma**2) ), color="c")
plt.xlabel('Smarts')
plt.ylabel('Probability density')
plt.title(r'Histogram of IQ: $\mu=100$, $\sigma=15$')

# Tweak spacing to prevent clipping of ylabel
plt.show()
```
