### numpy.random.random() ---- Generate random floating point numbers
The default is to generate a random floating point number, the range is between 0.0~1.0, you can also set the size of the returned data by the 
parameter size setting;

Generate a random floating point number:

```python
import numpy
n = numpy.random.random()
print (n)
```

Set the size of the returned data by the parameter size setting:
numpy.random.random(size=(3, 2))  -- means the resule is 3 rows and 2 columns. 

```python
import numpy
n = numpy.random.random(size=(3, 2))
print (n)
```

### numpy.random.normal()  Gaussian distribution random number

API: normal(loc=0.0, scale=1.0, size=None)

The probability density for the Gaussian distribution is

$$p(x) = \frac{1}{\sqrt{ 2 \pi \sigma^2 }}
e^{ - \frac{ (x - \mu)^2 } {2 \sigma^2} }$$,

where $\mu$ is the mean and $\sigma$ the standard deviation. The square of the standard deviation, $\sigma^2$, is called the variance.
The function has its peak at the mean, and its â€œspreadâ€ increases with the standard deviation (the function reaches 0.607 times its maximum at $x + \sigma$ and $x - \sigma$). This implies that [numpy.random.normal][1] is more likely to return samples lying close to the mean, rather than those far away.

[1]: https://docs.scipy.org/doc/numpy/reference/generated/numpy.random.normal.html#numpy.random.normal

```python
import numpy
n = numpy.random.normal(loc=1, scale=9, size=1)
print (n)
```
